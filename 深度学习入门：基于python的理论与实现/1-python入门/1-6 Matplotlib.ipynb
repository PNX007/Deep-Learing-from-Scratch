# 1-6-1 绘制简单图形

可以使用matplotlib的pyplot模块绘制图形，下面来看绘制sin函数的例子


```python
import numpy as np
import matplotlib.pyplot as plt

# 生成数据

x = np.arange(0, 10, 0.1)  # 以0.1位单位，生成0到6的数据
print("x = ",x)
y = np.sin(x)
print("y = ",y)
z = np.cos(x)
# 绘制图形

plt.plot(x, y)
plt.plot(x, z)
plt.plot(y, z)
plt.show()
```

# 1-6-2 pyplot的功能

在刚才的图形中，我们尝试使用pyplot的添加标题和x轴标签名等其他功能


```python
import numpy as np
import matplotlib.pyplot as plt

# 生成数据
x = np.arange(0, 10, 0.1)  # 以0.1位单位，生成0到6的数据
print("x = ",x)
y1 = np.sin(x)
print("y1 = ",y1)
y2 = np.cos(x)
print("y2 = ",y2)

# 绘制图形
plt.plot(x, y1, label = "sin")
plt.plot(x, y2, linestyle = "--", label = "cos")  # 用虚线绘制图形
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin &cos")
plt.legend() # what's mean this?
plt.show()

```

# 1-6-3 显示图像

pyplot中还提供了用于显示图像的方法（imshow（））。另外可以使用matplotlib.image模块的imread（）方法读入图像。下面来看例子。

```python
import matplotlib.pyplot as plt
from matplotlib.image import imread
img = imread("教育.png")
plt.imshow(img)

plt.show()
```
